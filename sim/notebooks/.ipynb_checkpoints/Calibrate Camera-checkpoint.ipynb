```python

import pandas as pd
import numpy as np
import cv2, os, copy
from sklearn import linear_model
import statsmodels.api as sm
from math import cos, radians, sin, degrees, pi, acos, sqrt, atan
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

class Video(object):
    
    def __init__(self, filename, rect=None, zoom=0, frame_rate=240,
                 distance=50, s=0):      

        self.num_frames=0
        self.distance=distance
        self.frame_rate = frame_rate
        self.s=s
        self.frames = []
        # bring in the frames --  set rectangle default with first frame
        vc = cv2.VideoCapture(filename)
        if vc.isOpened():
            rval , frame = vc.read()
            self.rect = ((0,0), (frame.shape[0:2])) if rect is None else rect
        else:
            rval = False
        
        while rval:
            self.num_frames += 1
            self.frames.append(frame)
            rval, frame = vc.read()
            cv2.waitKey(1)
```
