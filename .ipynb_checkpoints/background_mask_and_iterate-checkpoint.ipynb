```python
class Frame(object):
    
    def __init__(self, contours):
        self.contours = contours
        

class ContourBlob(object):
    
    def __init__(self, contour):
        self.moments = cv2.moments(contour)
        self.area = cv2.contourArea(contour)
        if moments1['m00'] > 0 and moments2['m00'] > 0:
            self.x_cord = self.moments['m10'] / self.moments['m00']
            self.y_cord = self.moments['m01'] / self.moments['m00']
        else:
            self.x_cord = None
            self.y_cord = None

    def distance(other_contour):
        if self.x_cord is not None and self.y_cord is not None and other_contour.x_cord is not None and other_contour.y_cord is not None:
            return sqrt((self.x_cord-other_contour.x_cord)**2 + (self.y_cord-other_contour.y_cord)**2)
        else:
            return None
```

```python
def findCentroid(contour):
    M = cv2.moments(contour)
    try:
        cx = int(M['m10']/M['m00'])
        cy = int(M['m01']/M['m00'])
        return (cx, cy)
    except ZeroDivisionError:
        return None
```
